```python
tickers = ['IBM', 'AAPL', 'MSFT']

def get_data(ticker,startdate, enddate):
    def data(ticker):
        return pdr.get_data_google(ticker, start = startdate, end = enddate)
    datas = map (data, tickers)
    return pd.concat(datas, keys = tickers, names = ['ticker', 'Date'])
```

```python
all_data = get_data(tickers, '2006-06-01', '2013-06-01')
```

```python
all_data.head(3)
```

```python
all_data.loc['IBM'].describe()
```

```python
all_data.loc['AAPL'].describe()
```

```python
daily_close_px = all_data['Close'].loc['IBM']
```
